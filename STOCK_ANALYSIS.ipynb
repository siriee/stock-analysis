# **Return on Investment**

Return on Investment (TATA MOTORS)

```python
#tata motors(if we buy one share of each stock on the 30th of each month beginning from January 2000 for Tata Motors)
sumtm = 0 #total amount invested in tata motors
stm=0     #total share bought in tata motor  
for i in range(len(motors)):
  if motors.loc[i,'Day']==30:
    sumtm+=motors.loc[i,'Open']
    stm+=1
print("total amount invested in tata motors shares are ",sumtm)
print("total number of stocks bought are ",stm)
print("average investment on each stock",(sumtm/stm))
```

```python
motors.tail(1)
```

```python
tm_end=298.2
result1=(tm_end*stm)-sumtm
print("Investment result:")
if result1 < 0:
  print("the net urealized loss:",result1)
else:
  print("the net unrealized profit:",result1)
roitm=(result1/sumtm)*100
print("Tata Motors ROI from 2000-1-3 to 2021-04-30 =",roitm,"%")    
```

Return on Investment(TATA STEEL)

```python
#tata steel(if we buy one share of each stock on the 30th of each month beginning from January 2000 for Tata steel)
sumts = 0 #total amount invested in tata steel
sts=0     #total share bought in tata steel  
for i in range(len(steel)):
  if steel.loc[i,'Day']==30:
    sumts+=steel.loc[i,'Open']
    sts+=1
print("total amount invested in tata steel shares are ",sumts)
print("total number of stocks bought are ",sts)
print("average investment on each stock",(sumts/sts))
```

```python
steel.tail(1)
```

```python
tsend=1024.0
result2=(tsend*sts)-sumts
print("Investment result:")
if result2 < 0:
  print("the net unrealized loss:",result2)
else:
  print("the net unrealized profit:",result2)
roits=(result2/sumts)*100
print("Tata steel ROI from 2000-1-3 to 2021-04-30 =",roits,"%")    
```

Return on Investment(TCS)

```python
#tata consultancy services(if we buy one share of each stock on the 30th of each month beginning from January 2000 for Tata steel)
sumtcs = 0 #total amount invested in tata consultancy service
stcs=0     #total share bought in tata consultancy service
for i in range(len(tcs)):
  if tcs.loc[i,'Day']==30:
    sumtcs+=tcs.loc[i,'Open']
    stcs+=1
print("total amount invested in tata consultancy services shares are ",sumtcs)
print("total number of stocks bought are ",stcs)
print("average investment on each stock",(sumtcs/stcs))
```

```python
tcs.tail(1)
```

```python
tcsend=3099.0
result3=(tcsend*stcs)-sumtcs
print("Investment result:")
if result3 < 0:
  print("the net unrealiazed loss:",result3)
else:
  print("the net unrealiazed profit:",result3)
roitcs=(result3/sumtcs)*100
print("Tata Consultancy Service ROI from 2004-08-25 to 2021-04-30 =",roitcs,"%")    
```

# **Investment results (Graphically)**

plotting ROI on bar graph

```python
stock=['tata motors','tata steel','tcs']
result=[roitm,roits,roitcs]
plt.bar(stock,result,color=['red','blue','grey'])
plt.title("ROI")
plt.xlabel("companies")
plt.ylabel("percentage")
plt.show()
```

plotting the loss or profit on bar graph

```python
stock=['tata motors','tata steel','tcs']
result=[result1,result2,result3]
plt.bar(stock,result,color=['black','blue','green'])
plt.title("Profit or loss")
plt.xlabel("companies")
plt.ylabel("amount")
plt.show()
```

```python
shares=[stm,sts,stcs]

plt.pie(shares,labels=['tata motors','tata steel','tcs'],autopct='%1.2f%%',colors=['violet','pink','yellow'])
plt.title('portfolio allocation')
plt.legend(title='',loc='upper left')
plt.show()
```
